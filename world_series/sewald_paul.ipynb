```python
import pandas as pd

df = pd.read_csv('623149_data.csv')
df
```

```python
df = df.sort_values(by = ['game_date','at_bat_number','pitch_number'], ascending=[True,True,True]).reset_index(drop=True)

uq_pitches = ['FF', 'ST']
first_pitch = []
first_zone = []
bat_count = []
this_batter = 0
count = 0
for pitch in df.index:
    if df['batter'][pitch] != this_batter:
        bat_count.append(count)
        count = 0
        this_batter = df['batter'][pitch]
        first_pitch.append(df['pitch_type'][pitch])
        first_zone.append(df['zone'][pitch])
        count += 1
    elif df['batter'][pitch] == this_batter:
        count += 1

uq_zones = list(set(first_zone))

for pitch in uq_pitches:
    print(pitch, first_pitch.count(pitch) / len(first_pitch), first_pitch.count(pitch))
    
print('')
for zone in uq_zones:
    print(zone, first_zone.count(zone) / len(first_zone), first_zone.count(zone))
    
sum(bat_count) / len(bat_count)
```
